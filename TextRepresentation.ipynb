```
#drop the unnecessary delimiters/punctuation
s = "@\"\")\,~'\'?%/$-&_;.!|\n:(#\\[]"
for i in range(len(df)):
  tmp = df['postText'][i]
  for obj in s:
    if obj in tmp:
      tmp = tmp.replace(obj,'') #'' instead of ' '
      df.set_value(i,'postText',tmp)
```

```
for i in range(len(df)):
  print(df['postText'][i])
```

```
def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
```

```
for i in range(len(df)):
  emofree = deEmojify(df['postText'][i])
  df.set_value(i, 'postText', emofree)
```

```
#remove string 'RT'
for i in range(len(df)):
  tmp = df['postText'][i]
  if 'RT' in tmp:
    tmp = tmp.replace('RT','')
    df.set_value(i,'postText',tmp)
```

```
#postMedia modification
x = ']['
for i in range(len(df)):
  tmp = df['postMedia'][i]
  for brac in x:
    tmp = tmp.replace(brac,'')
    df.set_value(i,'postMedia',tmp)
```

```
#before stemming
df.to_csv(r'/content/Drive/My Drive/Spring 2019/ML/sample for project/after_stem.csv', index= False)
```

Stemming the words

```
import nltk
nltk.download('punkt')
```

```
#stemming words using NLTK
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```
ps = PorterStemmer()

for i in range(len(df)):
  tmp = df['postText'][i]
  words = word_tokenize(tmp)
  for w in words:
    tmp = tmp.replace(w, ps.stem(w))
    df.set_value(i,'postText',tmp)
```
